```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#Use matlotlib to display the image
plt.imshow(img)
plt.title('ORIGINAL IMAGE')
plt.show()
```

```python
#Adjust to gray image
gray_img = cv2.imread('/content/1.jpg')
gray_img = cv2.cvtColor(gray_img, cv2.COLOR_BGR2GRAY)

#Show gray image
plt.imshow(gray_img, cmap = 'gray')
plt.title('GRAY IMAGE')
plt.show()
```

```python
#Adjust to Brightness image
brt_img = img.astype(np.float32) + 50
brt_img = np.clip(brt_img, 0, 255)
brt_img = brt_img.astype(np.uint8)

#Show brightness image
plt.imshow(brt_img)
plt.title('BRIGHTNESS IMAGE')
plt.show()
```

```python
#Adjust to Darkness image
dark_img = img.astype(np.float32) - 80
dark_img = np.clip(dark_img, 0, 255)
dark_img = dark_img.astype(np.uint8)

#Show brightness image
plt.imshow(dark_img)
plt.title('DARKNESS IMAGE')
plt.show()
```
