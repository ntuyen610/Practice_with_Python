- Bộ lọc Gauss: kỹ thuật xử lí ảnh dùng để làm mịn ảnh và khử nhiễu. Nguyên lí: áp dụng ma trận Gaus lên từng pixel ảnh
- OpenCV cung cấp 2 hàm để thực hiện Gauss:
- cv2.GaussianBlur: cách đơn giản để áp dụng bộ lọc  
- cv.filter2D: áp dụng một ma trận tuỳ ý lên ảnh

```python
import cv2
import matplotlib.pyplot as plt

#Load image
img = cv2.imread('/content/2.jpg', 0)

#Apply Gaussian Filter with kernel matrux 5x5, sigma = 1
img_filtered = cv2.GaussianBlur(img, (5,5), 1)
plt.title("Origin Image")
plt.imshow(img_filtered, cmap ="gray")
```

```python
img_filtered_2 = cv2.GaussianBlur(img, (5,5), 2.5)
plt.title("Sigma = 2.5")
plt.imshow(img_filtered_2, cmap ="gray")
```

```python
import numpy as np
import math

def gaussian_kernel(size, sigma):
  if size % 2 == 0:
    size = size + 1

  max_point = size // 2 #both directions (x,y) maximum cell start point
  min_point = -max_point #both directions(x,y) minimum cell start point

  K = np.zeros((size,size)) #kernel matrix
  for x in range(min_point, max_point + 1):
    for y in range(min_point, max_point +1):
      value = (1/(2 * math.pi * sigma**2) * math.exp((x**2 + y**2)/ (-2 * sigma**2)))
      K[x - min_point, y - min_point] = value

  return K
```

```python
kernel_5 = gaussian_kernel(5, 5)
img_gaussian_5 = cv2.filter2D(img, -1, kernel_5)
plt.title('Sigma = 5.0')
plt.imshow(img_gaussian_5, cmap ="gray")
plt.show()
```

```python
kernel_10 = gaussian_kernel(5, 10)
img_gaussian_10 = cv2.filter2D(img, -1, kernel_10)
plt.title('Sigma = 5.0')
plt.imshow(img_gaussian_10, cmap ="gray")
plt.show()
```
