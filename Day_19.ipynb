```python
lst_data = list(range(1,11))
print(lst_data)
```

```python
with open("data.txt", 'w') as f:
  f.write('-'.join(str(item) for item in lst_data))
```

```python
with open("data.txt", 'r') as f2:
  df = f2.read().split("-")
  num_lst = [int(item) for item in df]
  filter_lst = [i for i in num_lst if i % 3 == 0]
  print(filter_lst)
```
